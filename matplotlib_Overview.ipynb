## Making figures with NumPy arrays

```python
import numpy as np
```

```python
# Create some data

x = np.linspace(0,10,100)
x[:10]
```

```python
# Plot the data and create a line plot

fig, ax = plt.subplots()
ax.plot(x,x**2);
```

```python
# Use same data to make a scatter plot
```

```python
fig, ax = plt.subplots()
ax.scatter(x,np.exp(x));
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# Make a plot from dictionary
nut_butter_prices = {"Almond butter": 10,
                    "Peanut butter": 8,
                    "Cashew butter": 12}

fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title = "Nut butter store",
      ylabel = "Price ($)");
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
```

```python
# Make some data
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

```python
# Subplot option 1

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2,
                                            ncols = 2,
                                            figsize = (10,5))

# plot to each different axis
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplot option 2
fig, ax = plt.subplots(nrows = 2,
                      ncols = 2,
                      figsize = (10,10))
# Plot to each different index
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

```python
car_sales
```

```python
# replace dollar sign and comma with an empty string to perform numerical operations 
car_sales["Price"] = car_sales["Price"].str.replace('[\$\,\.]', '')
car_sales
```

```python
# remove last two zeros

car_sales["Price"] = car_sales["Price"].str[:-2]
car_sales
```

```python
car_sales["Sale Date"] = pd.date_range("1/1/2020", periods = len(car_sales))
car_sales
```

```python
# create a new column total sales
type(car_sales["Price"][0])
```

```python
car_sales["Total Sales"] = car_sales["Price"].astype(int).cumsum()
car_sales
```

```python
# plot total sales
car_sales.plot(x = "Sale Date", y = "Total Sales");
```

```python
# Reassign price column to int

car_sales["Price"] = car_sales["Price"].astype(int)
car_sales.plot(x = "Odometer (KM)", y = "Price", kind = "scatter");
```
